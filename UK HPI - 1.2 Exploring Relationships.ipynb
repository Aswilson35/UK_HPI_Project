## Variables to use:
â€¢	Index - House price index for a geography in a particular period (January 2015=100)
â€¢	DetachedIndex â€“ House price index for detached house for a geography in a particular period (January 2015=100)
â€¢	SemiDetachedPrice - Average house price for a semi-detached house for a geography in a particular period. 
â€¢	TerracedIndex â€“ House price index for a terraced house for a geography in a particular period (January 2015=100)
â€¢	FlatIndex - House price index for a flat house for a geography in a particular period (January 2015=100)
â€¢	FTBIndex â€“ House price index by buyer status (such as first time buyer) for a geography in a particular period (January 2015=100)
â€¢	FTBPrice â€“ Average house price by buyer status (such as first time buyer) for a geography in a particular period. 
â€¢	FOOPrice â€“ House price index by buyer status (such as former owner occupier) for a geography in a particular period.
â€¢	NewSalesVolume â€“ Number of registered transactions New for a geography in a a particular period. 
â€¢	OldSalesVolume â€“ Number of registered transactions Old for a geography in a particular period. 
â€¢	Official Interest Rate â€“ Bank of England official interest rate at time of observation. 
â€¢	MortgagePrice â€“ Average house price by funding status (such as mortgage) in a particular period. 
â€¢	CashPrice â€“ Average house price by funding status (such as cash) for a geography in a particular period.

Index â€“ FOOIndex = 0.87 medium â€“ high strength positive relationship, as the average house index increases so too does the former owner occupier (FOO) index.

Index â€“ FOOPrice = 0.75 medium â€“ medium positive relationship, as the average house price index increases so too does the former owner occupier (FOO) house price.
